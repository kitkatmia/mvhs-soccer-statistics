```python
def plot_possession_graphs(top_time_against):
    fig, axes = plt.subplots(2, 2, figsize=(20, 12), sharex=True)
    fig.patch.set_facecolor('white')

    axes[0, 0].plot(df['time_frame'], df['pass_success_rate'],
                    label="Recent LA Game", marker='o', color='b')
    axes[0, 0].plot(df_past['time_frame'], df_past['pass_success_rate'],
                    label="Past LA Game", marker='o', color='gray', alpha=0.5)
    axes[0, 0].set_ylabel('Pass Success Rate (%)')
    axes[0, 0].set_title('Time vs Pass Success Rate')

    axes[1, 0].plot(df['time_frame'], df['num_pass_strings'],
                    marker='o', color='b')
    axes[1, 0].plot(df_past['time_frame'], df_past['num_pass_strings'],
                    marker='o', color='gray', alpha=0.5)
    axes[1, 0].set_title('Time vs Number of Pass Strings')
    axes[1, 0].set_ylabel('Num of Pass Strings')

    axes[0, 1].plot(df['time_frame'], df['avg_pass_string'],
                    marker='o', color='b')
    axes[0, 1].plot(df_past['time_frame'], df_past['avg_pass_string'],
                    marker='o', color='gray', alpha=0.5)
    axes[0, 1].set_title('Time vs Average Passes in Pass String')
    axes[0, 1].set_ylabel('Average Passes in Pass String')

    axes[1, 1].plot(df['time_frame'], df['poss_%'], marker='o', color='b')
    axes[1, 1].plot(df_past['time_frame'], df_past['poss_%'],
                    marker='o', color='gray', alpha=0.5)
    axes[1, 1].set_title('Time vs Possession %')
    axes[1, 1].set_ylabel('Possession %')

    if (type(top_time_against) == list):
        for i in range(0, 2):
            for j in range(0, 2):
                if (i != 0 or j != 0):
                    axes[i, j].axvspan(top_time_against[0], top_time_against[1], color="#ff4040",
                                       alpha=0.5)
        axes[0, 0].axvspan(top_time_against[0], top_time_against[1], color="#ff4040",
                           alpha=0.5, label="End of game energy dip")

    for i in range(0, 2):
        for j in range(0, 2):
            if (i != 0):
                # to make sure there are shared x-labels
                axes[i, j].set_xlabel('Time Interval in Game (mins)')
    # axes[i, j].legend(loc='lower left')
    # fig.legend(loc="lower left")

    fig.legend(loc='upper center', bbox_to_anchor=(0.5, 0.93),
               ncol=3, fancybox=True, shadow=True)
    fig.suptitle("Possession Graph Comparisons for both LA Games", fontsize=28)
    plt.tight_layout(rect=[0, 0.03, 1, 0.93])
    plt.show()


plot_possession_graphs(["61-75", "76-90"])
```

```python
import seaborn as sns
# Set the 'timeFrame' column as the index for better plotting
df_copy = df.drop(columns=["shots"], inplace=False).set_index(
    'time_frame', inplace=False)
df_past_copy = df_past.set_index('time_frame', inplace=False)

# Draw line plots for each column against timeFrame
# plt.figure(figsize=(10, 12))

# Calculate the number of rows needed for subplots
num_features = len(df_copy.columns)
num_rows = (num_features + 1) // 2  # Ensure at least 1 row

# Create subplots with two plots per row
fig, axes = plt.subplots(num_rows, 2, figsize=(12, num_rows * 4))

# Flatten the axes array for ease of indexing
axes = axes.flatten()

for i, column in enumerate(df_copy.columns):
    sns.lineplot(data=df_copy[column], marker='o',
                 color='b', label="new", ax=axes[i])
    sns.lineplot(data=df_past_copy[column], marker='o',
                 color='gray', label="old", ax=axes[i])
    axes[i].set_title(f'{column} Over Time')
    axes[i].set_xlabel('Time Frame')
    axes[i].set_ylabel('Values')
    axes[i].legend()

# Remove empty subplots if there are an odd number of features
if num_features % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

## Observations

- We took more shots in the recent LA game (more chances on goal)
  - Confirmed by higher avg. of pass attemtps in atk + higher overall success
    - Also by more crosses!
- We had relative success in the defense
  - High pass success rate in intervals where they scored suggests that goals were break throughs on their part (e.g. we didn't get a chance to regain possession and pass it)

```python
import matplotlib.patches as mpatches


def plot_individual_pass_graphs():
    fig, axes = plt.subplots(3, 2, figsize=(20, 15), sharex=True)
    # set background to white
    fig.patch.set_facecolor('white')

    axes[0, 0].plot(df['time_frame'], df['pass_attempts_atk'],
                    label="Recent LA Game", marker='o', color='b')
    axes[0, 0].plot(df_past['time_frame'], df_past['pass_attempts_atk'],
                    label="Past LA Game", marker='o', color='gray', alpha=0.5)
    axes[0, 0].set_ylabel('PA - Attacking Third')
    axes[0, 0].set_title('Time vs Pass Atempts in the Attacking Third')

    axes[0, 1].plot(df['time_frame'], df['atk_third_pass_success_%'],
                    marker='o', color='b')
    axes[0, 1].plot(df_past['time_frame'], df_past['atk_third_pass_success_%'],
                    marker='o', color='gray', alpha=0.5)
    axes[0, 1].set_title('Time vs Pass Success Rate in Attacking Third')
    axes[0, 1].set_ylabel('PSR in Attacking Third (%)')

    axes[1, 0].plot(df['time_frame'], df['pass_attempts_mid'],
                    marker='o', color='b')
    axes[1, 0].plot(df_past['time_frame'], df_past['pass_attempts_mid'],
                    marker='o', color='gray', alpha=0.5)
    axes[1, 0].set_title('Time vs Pass Attempts - Mid Third')
    axes[1, 0].set_ylabel('PA - Mid Third')

    axes[1, 1].plot(df['time_frame'],
                    df['mid_third_pass_success_%'], marker='o', color='b')
    axes[1, 1].plot(df_past['time_frame'], df_past['mid_third_pass_success_%'],
                    marker='o', color='gray', alpha=0.5)
    axes[1, 1].set_title('Time vs Pass Success Rate in Mid Third')
    axes[1, 1].set_ylabel('PSR in Mid Third (%)')

    axes[2, 0].plot(df['time_frame'], df['pass_attempts_def'],
                    marker='o', color='b')
    axes[2, 0].plot(df_past['time_frame'], df_past['pass_attempts_def'],
                    marker='o', color='gray', alpha=0.5)
    axes[2, 0].set_title('Time vs Pass Attempts - Def Third')
    axes[2, 0].set_ylabel('PA - Def Third')

    axes[2, 1].plot(df['time_frame'],
                    df['def_third_pass_success_%'], marker='o', color='b')
    axes[2, 1].plot(df_past['time_frame'], df_past['def_third_pass_success_%'],
                    marker='o', color='gray', alpha=0.5)
    axes[2, 1].set_title('Time vs Pass Success Rate in Def Third')
    axes[2, 1].set_ylabel('PSR in Def Third (%)')

    # x-labels
    axes[2, 0].set_xlabel('Time Interval in Game (mins)')
    axes[2, 1].set_xlabel('Time Interval in Game (mins)')

    # title
    fig.suptitle("Specific Pass Comparisons for both LA Games", fontsize=28)

    # add acronym explanation
    plt.plot([], [], alpha=0, label='PSR = Pass Success Rate')
    plt.plot([], [], alpha=0, label='PA = Pass Attempts')

    # legend + legend location
    fig.legend(loc='upper center', bbox_to_anchor=(0.5, 0.93),
               ncol=4, fancybox=True, shadow=True)

    plt.tight_layout(rect=[0, 0.03, 1, 0.93])
    plt.show()


plot_individual_pass_graphs()
```
